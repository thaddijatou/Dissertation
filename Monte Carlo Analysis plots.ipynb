```python
#SAIFI
import numpy as np
import plotly.graph_objects as go
import matplotlib.pyplot as plt

height = [5.934, 6.136, 8.744, 8.495, 7.914, 7.914]
bars = ('2020', '2021','2022','2023','2024','2025')
x_pos = np.arange(len(bars))
plt.bar(x_pos, height, color = 'blue', width = 0.5)

plt.title('SAIFI')

plt.xticks(x_pos, bars)


plt.show()
```
